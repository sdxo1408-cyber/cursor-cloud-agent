````python
import os
from pathlib import Path

import anthropic
from dotenv import load_dotenv

# Load API key from .env in the project root
project_root = Path.cwd()
if project_root.name == "notebooks":
    project_root = project_root.parent

load_dotenv(project_root / ".env")

api_key = os.getenv("ANTHROPIC_API_KEY")
if not api_key or api_key == "your-api-key-here":
    raise ValueError(
        "ANTHROPIC_API_KEY is missing. Create a .env file in the project root."
    )

MODEL = os.getenv("ANTHROPIC_MODEL", "claude-sonnet-4-20250514")
client = anthropic.Anthropic(api_key=api_key)

print(f"Ready to use model: {MODEL}")
````

## 1. Your first message

Send a single user message and print Claude's reply.

````python
response = client.messages.create(
    model=MODEL,
    max_tokens=256,
    messages=[
        {"role": "user", "content": "Explain Python list comprehensions in one short paragraph."}
    ],
)

print(response.content[0].text)
````

## 2. Streaming responses

Stream tokens as they arrive — useful for chat-style UIs.

````python
with client.messages.stream(
    model=MODEL,
    max_tokens=256,
    messages=[
        {"role": "user", "content": "Write a tiny Python function that checks if a number is prime."}
    ],
) as stream:
    for text in stream.text_stream:
        print(text, end="", flush=True)

print()
````

## 4. System prompts

Use a `system` prompt to steer Claude's behavior.

````python
response = client.messages.create(
    model=MODEL,
    max_tokens=128,
    messages=[{"role": "user", "content": "Say hello in one sentence."}],
)

print("Input tokens:", response.usage.input_tokens)
print("Output tokens:", response.usage.output_tokens)
print("Reply:", response.content[0].text)
````

````python
messages = [
    {"role": "user", "content": "I'm learning Python. Suggest one beginner project."},
]

response = client.messages.create(model=MODEL, max_tokens=256, messages=messages)
assistant_reply = response.content[0].text
print("Assistant:", assistant_reply)

messages.append({"role": "assistant", "content": assistant_reply})
messages.append({"role": "user", "content": "Break that project into 3 small steps."})

follow_up = client.messages.create(model=MODEL, max_tokens=256, messages=messages)
print("\nAssistant:", follow_up.content[0].text)
````

````python
response = client.messages.create(
    model=MODEL,
    max_tokens=256,
    system="You are a patient Python tutor. Use simple examples and short answers.",
    messages=[
        {"role": "user", "content": "What is a dictionary in Python?"}
    ],
)

print(response.content[0].text)
````
